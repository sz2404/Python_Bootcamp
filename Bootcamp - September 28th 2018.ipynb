# Plotting

```python
import pylab as pl
%pylab inline
s = {'font.size':20.0, 'figure.figsize':[6.0, 6.0]}
pl.rcParams.update(s)
```

```python
pl.plot([0,1,2], [0,1,2], 'r*')
pl.xlabel('x')
pl.ylabel('y')
```

```python
pl.rcParams
```

```python
x = np.arange(100)
y = np.sqrt(x) + np.random.randn(100)
yerrs = np.sqrt(y)
```

```python
fig = pl.figure(figsize=(20,8), facecolor='gray')
#ax1 = fig.add_subplot(1,2,1) #1 row, 2 columns, 1st one
ax1 = fig.add_subplot(121)
ax1.plot(x,y,'o')
ax2 = fig.add_subplot(122)
#ax2.plot(x,y,'--', label='hi')
ax2.errorbar(x[:10], y[:10]+yerrs[:10], y[:10]-yerrs[:10], fmt='o', alpha = 0.5)
ax2.fill_between(x[:10], y[:10]+yerrs[:10], y[:10]-yerrs[:10], alpha = 0.5)
ax1.legend()
```

```python
myar = np.random.randn(200).reshape(20,10)
fig = pl.figure(figsize=(20,8), facecolor='gray')
#ax1 = fig.add_subplot(1,2,1) #1 row, 2 columns, 1st one
ax1 = fig.add_subplot(121)
ax1.imshow(myar, cmap='viridis', interpolation = 'bilinear')
ax1.axis('off')
```
